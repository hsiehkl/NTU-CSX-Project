```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
import numpy as np
% matplotlib inline
```

```python
# load file from given file name, will return a list of lines
def load_file(file):
    file =open(file+".txt", "r")
    content = file.readlines()
    return content
```

```python
ProperN = ["衍生性","放款及應收款","金融資產","持”’,”到期日","備供出售","金融負債","利息法","交易成本","確定承諾","預期交易","避險工具","被避險項目",
           "避險有效性","攤銷","貨幣性","營運機構”,”淨投資","或有負債","不動產","馬達","其他綜合損益","無形資產","製造","信託","董事長","新臺幣", 
           "新能效","員工"]
for w in ProperN:
    jieba.add_word(w)
```

```python
stopwords = ['我', '對', '\n', "(", ")", "100", "100.00%", " ", '-', '000', '1', "100%", '2', "0", "3"]
puncs = ',！？｡＂＃＄％＆＇（）＊＋，－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏.。'
```

```python
def get_word_dic(content):
    join_text = ''.join(content)
    
    seg_list = jieba.cut(join_text, cut_all=False)
    seg_list = list(seg_list)
    wordDict = {}
    for w in seg_list:
        if (w not in stopwords) and (w not in puncs):
            if w in wordDict:
                wordDict[w] = wordDict[w] + 1
            else:
                wordDict[w] = 1
    return wordDict
```

```python
def plot_cloud(wordDict):
    font = r"/anaconda3/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/SimHei.ttf"
    plt.style.use('ggplot')
    wordcloud = WordCloud(font_path=font)
    wordcloud.fit_words(wordDict)
    plt.figure(figsize=(12,10))
    #plt.title("參公司治理報告",fontsize=30)
    plt.grid(False)
    plt.imshow(wordcloud)
```

```python
def plot_bar(wordDict):
    frequent_key = sorted(wordDict, key=wordDict.get, reverse=True)
    first10_key = frequent_key[:10]
    first10_value = [wordDict[x] for x in first10_key]
    first30_key = frequent_key[:30]
    print("前30名文字"first30_key)
    N = 10
    keys = first10_key
    value = first10_value
    ind = np.arange(N)    # the x locations for the groups
    width = 0.4       # the width of the bars: can also be len(x) sequence
    p1 = plt.bar(ind, value, width)
    plt.ylabel('Frequency')
    plt.xlabel("Word")
    plt.title("Top 10")
    plt.xticks(ind, first10_key)
    plt.show()
```

```python
content3 = load_file("參公司治理報告")
word_dic3 = get_word_dic(content3)
```

```python
plot_cloud(word_dic3)
```

```python
plot_bar(word_dic3)
```
