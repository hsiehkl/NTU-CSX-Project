```python
import numpy as np
import pandas as pd
from climate2_getClimate import *
```

```python
df = pd.read_csv('../Orders.csv')
```

```python
#過濾nan，篩選台北市資料
df = df[df['City'] == df['City']]
df = df[df['City'].str.contains("台北市")]
```

```python
#確認city只剩台北市
df["City"].unique()
```

```python
df.head(5)
```

```python
# 取出 "SalesOrderSlaveDateTime、City、SalesOrderSlaveTotalPayment"
taipei_data = df[['SalesOrderSlaveDateTime', 'City', 'SalesOrderSlaveTotalPayment']]
```

```python
taipei_data['SalesOrderSlaveDateTime'] = pd.to_datetime(taipei_data['SalesOrderSlaveDateTime']).apply(lambda x: x.date()).astype(str)
```

```python
#data set裡有1073不同天的日期
dates = taipei_data["SalesOrderSlaveDateTime"].unique()
len(dates)
```

```python
#時間關係，只先爬後面30個日期
last_30_dates = dates[1043:]
last_30_dates
```

```python
taipei_precp = list()
taipei_temperature = list()
for i in range(len(last_30_dates)):
    print(last_30_dates[i])
    result = getClimate("臺北市", last_30_dates[i])
    taipei_precp.append(result[0])
    taipei_temperature.append(result[1])
```

```python
print(len(last_30_dates))
print(len(taipei_precp))
print(len(taipei_temperature))
```

```python
taipei_climate = pd.DataFrame({'SalesOrderSlaveDateTime': last_30_dates, 'Precp': taipei_precp, 'Temperature': taipei_temperature})
```

```python
taipei_climate
```

```python
taipei_merge = pd.merge(taipei_data, taipei_climate, on='SalesOrderSlaveDateTime')
```
